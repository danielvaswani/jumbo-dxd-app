```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('recipes9.csv')

df.head()
```

```python
df.plot(kind = 'scatter', x = 'PREPARATION_TIME', y = 'WAITING_TIME', color='gold')

plt.show() 
```

```python
plt.hist(df['TOTAL_TIME'], range=[0, 190], color='gold')
plt.ylabel("Total Time")
plt.xlabel("Amount")

plt.show()
```

```python
#AVERAGE CALORIE COUNT
avg_cal = df["kcal"].mean()
print(f"The average calorie count is {avg_cal} kcal")
```

```python
#AVERAGE RECIPE TIME
avg_time = df["TOTAL_TIME"].mean()
print(f"The average recipe time is {avg_time} min")
```

```python
df_filter_under_30 = df.loc[df['TOTAL_TIME'] < 30, ['TOTAL_TIME']]

num_under_30 = len(df_filter_under_30['TOTAL_TIME'])
print(f"The number of recipes under 30 min is {num_under_30}")


df_filter_over_30 = df.loc[df['TOTAL_TIME'] > 30, ['TOTAL_TIME']]

num_over_30 = len(df_filter_over_30['TOTAL_TIME'])
print(f"The number of recipes under 30 min is {num_over_30}")
```
